```python
import gcsfs
import xarray as xr
import timeout_decorator
```

```python
fs_gcs = gcsfs.GCSFileSystem(anon=True,)
```

```python
cache_base = 'gs://pangeo-forge-us-central1/pangeo-forge-cache'
recipe_cache = 'cesm_pop_lowres_1deg/v5_rel04_BC5_ne30_g16'
```

```python
for cached_file in fs_gcs.ls(f"{cache_base}/{recipe_cache}"):
    if "ssh_2" in cached_file:
        file = cached_file
```

```python
file = fs_gcs.open(file, mode="rb")
file
```

```python
@timeout_decorator.timeout(60)
def timed_dataset_opener(f):
    print(f"Attempting to open {f}")
    return xr.open_dataset(f)
```

```python
timed_dataset_opener(f=file)
```
